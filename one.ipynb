```python
import numpy as np
import pandas as pd 
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
data = pd.read_csv('1.01.+Simple+linear+regression.csv')
```

```python
data
```

```python
data.describe()
```

```python
y=data['GPA']
x1=data['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(x1,y)
yhat=0.0017*x1+0.275
fig=plt.plot(x1,yhat,lw=4,c="orange",label="regression line")
plt.xlabel("SAT",fontsize=20)
plt.ylabel("GPA",fontsize=20)
plt.show()
```

```python
import seaborn as sns
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
sns.set()
```

```python
dummy_data = pd.read_csv("1.03.+Dummies.csv")
```

```python
dummy_data
```

```python
data=dummy_data.copy()
data["Attendance"]=data["Attendance"].map({'Yes':1,'No':0})
```

```python
data.describe()
```

```python
y=data["GPA"]
x1=data[["SAT","Attendance"]]
```

```python
x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(data["SAT"],y)
yhat_no=0.6439+0.0014*x1
yhat_yes=0.8665+0.0014*x1
fig=plt.plot(data["SAT"],yhat_no,lw=1,c="red",label="regression line")
fig=plt.plot(data["SAT"],yhat_yes,lw=1,c="green",label="regression line")
plt.xlabel("SAT and Attendence",fontsize=20)
plt.ylabel("GPA",fontsize=20)
plt.show()
```
